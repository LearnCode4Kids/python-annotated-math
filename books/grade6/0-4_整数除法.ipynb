# 0.4 整数除法

整数除法回答两个问题：能分成几份？有没有剩下？在 Python 里，`//` 得到商，`%` 得到余数。

## 思维视角

**逆向思维 + 枚举验证。** 除法和乘法互为逆运算。检查除法结果时，可以用“商乘除数再加余数”验证。

## 商和余数

例如 `17 ÷ 5`，最多可以分出 3 组 5，还剩 2。

```python
number = 17
divisor = 5

quotient = number // divisor
remainder = number % divisor

print(quotient)
print(remainder)
print(divisor * quotient + remainder)
```

## 用减法模拟除法

如果暂时不用 `//`，也可以不断减去除数。能减几次，商就是几；最后剩下的数就是余数。

```python
number = 17
divisor = 5

left = number
quotient = 0

while left >= divisor:
    left = left - divisor
    quotient = quotient + 1

print("商:", quotient)
print("余数:", left)
```

## 整除

如果余数是 0，就说一个数能被另一个数整除。整除是后面学习因数、公约数的基础。

```python
for x in range(1, 13):
    if 12 % x == 0:
        print(x, "能整除 12")
```

## 变式练习

1. 用 `//` 和 `%` 计算 `29 ÷ 6` 的商和余数。
2. 不用 `//`，只用循环减法计算 `31 ÷ 4`。
3. 写一个函数 `divide_with_remainder(a, b)`，返回 `(商, 余数)`。
4. 用 `b * 商 + 余数 == a` 检查你的函数。

## 小结

整数除法有两个结果：商和余数。只要记住 `被除数 = 除数 × 商 + 余数`，就能检查每一步是否合理。
