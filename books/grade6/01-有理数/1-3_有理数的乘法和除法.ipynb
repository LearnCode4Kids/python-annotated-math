# 1.3 有理数的乘法和除法

## 学习目标

学完本节，你应该能做到：

- 判断两个有理数相乘时结果的符号。
- 计算有理数乘法：先定符号，再算绝对值。
- 理解多个有理数相乘时，负因数个数决定符号。
- 把有理数除法转化为乘法：$a \div b=a\times \frac{1}{b}$，其中 $b\ne 0$。

**思维视角：归纳 + 分类讨论 + 化归。**

归纳用来从例子里发现符号规律；分类讨论用来区分同号、异号和 0；化归用来把除法转成乘法。

## 有理数乘法法则

两个有理数相乘：

- 同号得正。
- 异号得负。
- 任何数与 0 相乘，积为 0。
- 绝对值部分直接相乘。

简单说：**先看符号，再乘绝对值。**

```python
# 约定：先只用 * 算正数和 0 的乘法。


def multiply(a, b):
    """计算两个整数相乘。"""
    if a == 0 or b == 0:
        return 0

    # 先算绝对值
    product = abs(a) * abs(b)

    # 同号得正，异号得负
    if (a > 0 and b > 0) or (a < 0 and b < 0):
        return product
    return -product


examples = [
    (3, 4),
    (-3, -4),
    (3, -4),
    (-3, 4),
    (0, -4),
]

for a, b in examples:
    print(f"{a} × {b} = {multiply(a, b)}")
```

上面的例子可以归纳成一张符号表：

| 第一个因数 | 第二个因数 | 积的符号 |
|---|---|---|
| 正 | 正 | 正 |
| 负 | 负 | 正 |
| 正 | 负 | 负 |
| 负 | 正 | 负 |

负数乘负数为什么是正数，可以先把它理解为“反方向的反方向”。

## 多个有理数相乘

多个非零有理数相乘时，只需要数一数有几个负因数：

- 负因数个数是偶数，积为正。
- 负因数个数是奇数，积为负。
- 只要有一个因数是 0，积就是 0。

```python
groups = [
    [2, -3, -4],
    [-2, -3, -4],
    [5, -2, 0, -7],
]

for numbers in groups:
    product = 1
    negative_count = 0
    for x in numbers:
        product *= x
        if x < 0:
            negative_count += 1

    print(f"{numbers}：负因数有 {negative_count} 个，积为 {product}")
```

## 乘法交换律、结合律和分配律

有理数乘法仍然满足小学学过的运算律：

$$a\times b=b\times a$$

$$(a\times b)\times c=a\times(b\times c)$$

$$a\times(b+c)=a\times b+a\times c$$

它们的作用是让计算更方便。

```python
print("交换律：", 3 * (-5), (-5) * 3)
print("结合律：", (2 * (-3)) * 4, 2 * ((-3) * 4))
print("分配律：", -3 * (10 + 2), (-3) * 10 + (-3) * 2)
```

## 有理数除法：转成乘法

除法可以化归为乘法：

$$a\div b=a\times \frac{1}{b}\quad(b\ne 0)$$

这里 $\frac{1}{b}$ 叫做 $b$ 的倒数。

注意：**0 不能作除数**，所以 $b$ 不能等于 0。

```python
from fractions import Fraction

examples = [
    (12, -3),
    (-8, -4),
    (Fraction(3, 5), Fraction(-2, 3)),
]

for a, b in examples:
    reciprocal = Fraction(1, 1) / b
    result = a * reciprocal
    print(f"{a} ÷ ({b}) = {a} × ({reciprocal}) = {result}")
```

## 混合计算：先处理符号，再处理数值

看一个例子：

$$(-6)\times\frac{2}{3}\div(-4)$$

先把除法改成乘法：

$$(-6)\times\frac{2}{3}\times\left(-\frac{1}{4}\right)$$

两个负因数，结果为正；数值部分是 $6\times\frac{2}{3}\times\frac{1}{4}=1$。

```python
from fractions import Fraction

result = Fraction(-6, 1) * Fraction(2, 3) / Fraction(-4, 1)
print(result)
```
