## 有理数加法法则

加法分成三类情况：

1. **同号两数相加**：符号不变，绝对值相加。

2. **异号两数相加**：先比较绝对值，再决定结果符号。

3. **一个数与 0 相加**：直接返回另一个数。

```python
def add(a, b):
    """
    有理数的加法
    """
    if a>0 and b> 0:
        return a + b
    elif a < 0 and b < 0:
        return -(-a + -b)
    elif a > 0 and b < 0:
        if a > -b:
            return a - (-b)
        else:
            return -((-b)-a)
    elif a < 0 and b > 0:
        if -a > b:
            return -((-a) - b)
        else:
            return b - (-a)
    else:
        return a + b
    
# 测试案例4个
print(add(3, 5))   # 8
print(add(-3, -5)) # -8
print(add(3, -5))  # -2
print(add(-3, 5))  # 2
```

## 有理数减法：减去一个数，等于加上它的相反数

减法这部分也不重新讲背景，直接抓最重要的代码改写：

$$5-(-2)$$

$$5-(-2)=5+2=7$$

$$a-b=a+(-b)$$

所以写函数时，减法最稳定的做法通常不是单独再写一套规则，而是先转成加法。

```python
from fractions import Fraction


def subtract_as_add(a, b):
    """把 a-b 改写成 a+(-b)。"""
    return a + (-b)


examples = [
    (6, -6),
    (0, 9),
    (Fraction(-1, 2), Fraction(-1, 4)),
    (Fraction(-3, 2), Fraction(1, 3)),
]

for a, b in examples:
    print(f"{a} - ({b}) = {a} + ({-b}) = {subtract_as_add(a, b)}")
```

## 混合运算：先把减法都改成加法

多项混合运算时，最实用的套路很简单：先统一成加法，再分组。

$$-16+7-(-13)-14$$

第一步，先把减法改成加法：

$$-16+7+13+(-14)$$

第二步，再用交换律、结合律重排：

$$(-16)+(-14)+(7+13)=-30+20=-10$$

```python
terms = [-16, 7, 13, -14]
negative_group = [-16, -14]
positive_group = [7, 13]

print("全部项：", terms, "和为", sum(terms))
print("负数组：", negative_group, "和为", sum(negative_group))
print("正数组：", positive_group, "和为", sum(positive_group))
print("合并：", sum(negative_group) + sum(positive_group))
```
