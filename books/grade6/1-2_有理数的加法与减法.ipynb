```python
import os
os.environ.setdefault("MPLCONFIGDIR", "/tmp/matplotlib")
os.environ.setdefault("XDG_CACHE_HOME", "/tmp")

from fractions import Fraction
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["axes.unicode_minus"] = False

def sign_name(x):
    if x > 0:
        return "正数"
    if x < 0:
        return "负数"
    return "0"

def signed_text(x):
    """显示带符号的数，但把 0 显示成 0，而不是 +0。"""
    return "0" if x == 0 else f"{x:+}"

def add_rule(a, b):
    """用文字解释 a+b 属于哪一种加法情形。"""
    if a == 0 or b == 0:
        return "一个数与 0 相加，仍得这个数。"
    if (a > 0 and b > 0) or (a < 0 and b < 0):
        return "同号相加：取相同符号，并把绝对值相加。"
    if abs(a) == abs(b):
        return "异号相加且绝对值相等：和为 0。"
    return "异号相加且绝对值不等：取绝对值较大数的符号，并用较大绝对值减较小绝对值。"

def draw_addition(a, b, xmin=-8, xmax=8, title="addition on a number line"):
    """在数轴上画出从 0 出发，先走 a，再走 b，最后到 a+b。"""
    result = a + b
    axis_right = xmax + 0.35
    fig, ax = plt.subplots(figsize=(9, 2.2))
    ax.annotate("", xy=(axis_right, 0), xytext=(xmin, 0), arrowprops=dict(arrowstyle="->", color="black"))

    ticks = np.arange(np.ceil(xmin), np.floor(xmax) + 1)
    for t in ticks:
        ax.plot([t, t], [-0.05, 0.05], color="black")
        ax.text(t, -0.16, str(int(t)), ha="center", va="top", fontsize=9)

    ax.annotate("", xy=(float(a), 0.35), xytext=(0, 0.35), arrowprops=dict(arrowstyle="->", color="#1f77b4", linewidth=2))
    ax.annotate("", xy=(float(result), 0.62), xytext=(float(a), 0.62), arrowprops=dict(arrowstyle="->", color="#d62728", linewidth=2))
    ax.scatter([0, float(a), float(result)], [0, 0, 0], color=["black", "#1f77b4", "#d62728"], zorder=3)
    ax.text(0, -0.36, "0", ha="center")
    ax.text(float(a), -0.36, f"a={a}", ha="center", color="#1f77b4")
    ax.text(float(result), -0.36, f"a+b={result}", ha="center", color="#d62728")
    ax.text((0 + float(a)) / 2, 0.45, f"先走 {a}", ha="center", color="#1f77b4")
    ax.text((float(a) + float(result)) / 2, 0.72, f"再走 {b}", ha="center", color="#d62728")

    ax.set_xlim(xmin - 0.3, axis_right + 0.05)
    ax.set_ylim(-0.55, 1.0)
    ax.set_title(title)
    ax.axis("off")
    plt.show()
```

## 有理数加法法则

先用代码把教材表格中的几个月算出来。代码不会替我们理解数学，但它能帮我们把规律看得更整齐。

```python
months = [
    ("第一个月", 3, 2),
    ("第二个月", -2, -1),
    ("第三个月", -1, 3),
    ("第四个月", 2, -2),
    ("第五个月", -2, 1),
]

for name, first_half, second_half in months:
    total = first_half + second_half
    print(f"{name}: {signed_text(first_half)} + {signed_text(second_half)} = {signed_text(total)}，{add_rule(first_half, second_half)}")
```

可以把加法法则分成三类：

1. **同号两数相加**：符号不变，绝对值相加。
   
   例如 $(-12)+(-36)=-(12+36)=-48$。

2. **异号两数相加**：先比较谁离 0 更远。
   
   例如 $(-16)+5=-(16-5)=-11$。

3. **一个数与 0 相加**：仍得这个数。
   
   例如 $-\frac{5}{4}+0=-\frac{5}{4}$。

## 数轴理解：先走一段，再走一段

有理数加法可以看成在数轴上的移动：

- 正数：向右走。
- 负数：向左走。
- 加法：从当前位置继续走。

```python
draw_addition(15, -25, xmin=-12, xmax=18, title="Truck route: 15 + (-25)")
```

教材例题里，卡车先向东行驶 15 km，再向西行驶 25 km。

规定向东为正，算式是：

$$15+(-25)=-10$$

结果是 -10，表示最后停在出发点以西 10 km 处。

## 加法交换律和结合律

小学学过自然数和正分数的加法交换律、结合律。引入负数后，它们仍然成立：

$$a+b=b+a$$

$$(a+b)+c=a+(b+c)$$

这很有用。遇到多个数相加时，可以先把容易凑整、互为相反数、同分母或同小数位的数放在一起。

```python
numbers = [16, -9, 24]
print("原顺序：", numbers, "和为", sum(numbers))

better_order = [16, 24, -9]
print("换顺序：", better_order, "和为", sum(better_order))

fractions = [Fraction(17, 8), Fraction(-5, 8), Fraction(-5, 4), Fraction(1, 4)]
print("分数组：", fractions)
print("总和：", sum(fractions))
```

观察第二组分数：

$$\frac{17}{8}+\left(-\frac{5}{8}\right)+\left(-1\frac{1}{4}\right)+\frac{1}{4}$$

可以先算：

$$\frac{17}{8}+\left(-\frac{5}{8}\right)=\frac{12}{8}=\frac{3}{2}$$

$$-1\frac{1}{4}+\frac{1}{4}=-1$$

所以总和是 $\frac{3}{2}+(-1)=\frac{1}{2}$。

## 有理数减法：减去一个数，等于加上它的相反数

教材用温度引入：一天最高气温 $5^\circ C$，最低气温 $-2^\circ C$，温差是多少？

算式是：

$$5-(-2)$$

减法可以转化成加法：

$$5-(-2)=5+2=7$$

一般地：

$$a-b=a+(-b)$$

这叫**化归**：把还不熟悉的减法，转成已经掌握的加法。

```python
def subtract_as_add(a, b):
    """把 a-b 改写成 a+(-b)。"""
    return a + (-b)

examples = [
    (6, -6),
    (0, 9),
    (Fraction(-1, 2), Fraction(-1, 4)),
    (Fraction(-3, 2), Fraction(1, 3)),
]

for a, b in examples:
    print(f"{a} - ({b}) = {a} + ({-b}) = {subtract_as_add(a, b)}")
```

减法最容易错在两个地方：

- 看见 $-(-6)$ 时忘了变成 $+6$。
- 把 $a-b$ 错看成 $b-a$。

用数轴看，$a-b$ 可以理解成：从 $b$ 到 $a$ 相差多少。

## 混合运算：先把减法都改成加法

教材例题：

$$-16+7-(-13)-14$$

先把减法改成加法：

$$-16+7+13+(-14)$$

再用交换律、结合律：

$$(-16)+(-14)+(7+13)=-30+20=-10$$

```python
terms = [-16, 7, 13, -14]
negative_group = [-16, -14]
positive_group = [7, 13]

print("全部项：", terms, "和为", sum(terms))
print("负数组：", negative_group, "和为", sum(negative_group))
print("正数组：", positive_group, "和为", sum(positive_group))
print("合并：", sum(negative_group) + sum(positive_group))
```

<summary>参考思路</summary>

1. $(-13)+(-9)=-22$；$(-16)+5=-11$；$\frac{2}{5}+(-\frac{3}{5})=-\frac{1}{5}$。
