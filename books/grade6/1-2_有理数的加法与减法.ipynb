## 有理数加法法则

先用代码把教材表格中的几个月算出来。代码不会替我们理解数学，但它能帮我们把规律看得更整齐。

```python
months = [
    ("第一个月", 3, 2),
    ("第二个月", -2, -1),
    ("第三个月", -1, 3),
    ("第四个月", 2, -2),
    ("第五个月", -2, 1),
]

for name, first_half, second_half in months:
    total = first_half + second_half

    if first_half == 0 or second_half == 0:
        rule = "一个数与 0 相加，仍得这个数。"
    elif first_half * second_half > 0:
        rule = "同号相加：取相同符号，并把绝对值相加。"
    elif abs(first_half) == abs(second_half):
        rule = "异号相加且绝对值相等：和为 0。"
    else:
        rule = "异号相加且绝对值不等：取绝对值较大数的符号，并用较大绝对值减较小绝对值。"

    first_text = f"{first_half:+}"
    second_text = f"{second_half:+}"
    total_text = "0" if total == 0 else f"{total:+}"
    print(f"{name}: {first_text} + {second_text} = {total_text}，{rule}")
```

可以把加法法则分成三类：

1. **同号两数相加**：符号不变，绝对值相加。
   
   例如 $(-12)+(-36)=-(12+36)=-48$。

2. **异号两数相加**：先比较谁离 0 更远。
   
   例如 $(-16)+5=-(16-5)=-11$。

3. **一个数与 0 相加**：仍得这个数。
   
   例如 $-\frac{5}{4}+0=-\frac{5}{4}$。

## 数轴理解：先走一段，再走一段

有理数加法可以看成在数轴上的移动：

- 正数：向右走。
- 负数：向左走。
- 加法：从当前位置继续走。

教材例题里，卡车先向东行驶 15 km，再向西行驶 25 km。

规定向东为正，算式是：

$$15+(-25)=-10$$

结果是 -10，表示最后停在出发点以西 10 km 处。

## 有理数的加法
有理数的加法法则
（ １ ）同号两数相加，取原来加数的符号，并把绝对值相加．
（ ２ ）异号两数相加，绝对值相等时和为０ ；绝对值不相等时，取绝对
值较大的加数的符号，并用较大的绝对值减去较小的绝对值．
（ ３ ）任何一个数与０相加，仍得这个数．

```python
def positive_add(x, y):
    """返回两个非负数 x 和 y 的和。"""
    assert x >= 0 and y >= 0, "x 和 y 必须是非负数"
    return x + y


def positive_difference(larger, smaller):
    """返回两个非负数的差：larger - smaller。"""
    assert larger >= 0 and smaller >= 0, "两个数必须是非负数"
    assert larger >= smaller, "larger 必须不小于 smaller"
    return larger - smaller


def my_add(x, y):
    """按教材中的有理数加法法则计算 x + y。"""
    if x == 0:
        return y
    if y == 0:
        return x

    # 同号相加：符号不变，绝对值相加。
    if x > 0 and y > 0:
        return positive_add(x, y)
    if x < 0 and y < 0:
        return -positive_add(-x, -y)

    # 异号相加：比较绝对值，取绝对值较大数的符号。
    abs_x = abs(x)
    abs_y = abs(y)
    if abs_x == abs_y:
        return 0
    if abs_x > abs_y:
        difference = positive_difference(abs_x, abs_y)
        return difference if x > 0 else -difference

    difference = positive_difference(abs_y, abs_x)
    return difference if y > 0 else -difference


examples = [(3, 2), (-2, -1), (-1, 3), (2, -2), (-2, 1)]
for a, b in examples:
    result = my_add(a, b)
    result_text = "0" if result == 0 else f"{result:+}"
    print(f"{a:+} + {b:+} = {result_text}")
```

## 加法交换律：换顺序，和不变

小学学过自然数和正分数的加法交换律。引入负数后，它仍然成立：

$$a+b=b+a$$

直观地说，两个数相加时，谁先出现、谁后出现，不影响最后的和。比如先算 $16+(-9)$，或者先算 $(-9)+16$，结果一样。


```python
a, b = 16, -9

print("交换律：")
print(f"{a} + ({b}) = {a + b}")
print(f"({b}) + {a} = {b + a}")
```

## 加法结合律：换括号，和不变

三个有理数相加时，先把前两个数相加再与第三个数相加，或者先把后两个数相加再与第一个数相加，和不变：

$$(a+b)+c=a+(b+c)$$

括号只是在告诉我们“先算哪一部分”。结合律说明：只要都是加法，改变括号不会改变总和。


```python
x, y, z = 16, -9, 24

print("结合律：")
print(f"({x} + ({y})) + {z} = {(x + y) + z}")
print(f"{x} + (({y}) + {z}) = {x + (y + z)}")
```

交换律和结合律很有用。遇到多个数相加时，可以先把容易凑整、互为相反数、同分母或同小数位的数放在一起。


```python
from fractions import Fraction

numbers = [16, -9, 24]
print("实际计算时，可以先凑整：")
print(f"16 + (-9) + 24 = {sum(numbers)}")
print(f"16 + 24 + (-9) = {16 + 24 + (-9)}")

fractions = [Fraction(17, 8), Fraction(-5, 8), Fraction(-5, 4), Fraction(1, 4)]
print()
print("分数也一样：")
print(f"17/8 + (-5/8) + (-5/4) + 1/4 = {sum(fractions)}")
print(f"(17/8 + (-5/8)) + ((-5/4) + 1/4) = {(fractions[0] + fractions[1]) + (fractions[2] + fractions[3])}")
```

观察第二组分数：

$$\frac{17}{8}+\left(-\frac{5}{8}\right)+\left(-1\frac{1}{4}\right)+\frac{1}{4}$$

可以先算：

$$\frac{17}{8}+\left(-\frac{5}{8}\right)=\frac{12}{8}=\frac{3}{2}$$

$$-1\frac{1}{4}+\frac{1}{4}=-1$$

所以总和是 $\frac{3}{2}+(-1)=\frac{1}{2}$。

## 有理数减法：减去一个数，等于加上它的相反数

教材用温度引入：一天最高气温 $5^\circ C$，最低气温 $-2^\circ C$，温差是多少？

算式是：

$$5-(-2)$$

减法可以转化成加法：

$$5-(-2)=5+2=7$$

一般地：

$$a-b=a+(-b)$$

这叫**化归**：把还不熟悉的减法，转成已经掌握的加法。

```python
from fractions import Fraction


def subtract_as_add(a, b):
    """把 a-b 改写成 a+(-b)。"""
    return a + (-b)


examples = [
    (6, -6),
    (0, 9),
    (Fraction(-1, 2), Fraction(-1, 4)),
    (Fraction(-3, 2), Fraction(1, 3)),
]

for a, b in examples:
    print(f"{a} - ({b}) = {a} + ({-b}) = {subtract_as_add(a, b)}")
```

减法最容易错在两个地方：

- 看见 $-(-6)$ 时忘了变成 $+6$。
- 把 $a-b$ 错看成 $b-a$。

用数轴看，$a-b$ 可以理解成：从 $b$ 到 $a$ 相差多少。

## 混合运算：先把减法都改成加法

教材例题：

$$-16+7-(-13)-14$$

先把减法改成加法：

$$-16+7+13+(-14)$$

再用交换律、结合律：

$$(-16)+(-14)+(7+13)=-30+20=-10$$

```python
terms = [-16, 7, 13, -14]
negative_group = [-16, -14]
positive_group = [7, 13]

print("全部项：", terms, "和为", sum(terms))
print("负数组：", negative_group, "和为", sum(negative_group))
print("正数组：", positive_group, "和为", sum(positive_group))
print("合并：", sum(negative_group) + sum(positive_group))
```

<summary>参考思路</summary>

1. $(-13)+(-9)=-22$；$(-16)+5=-11$；$\frac{2}{5}+(-\frac{3}{5})=-\frac{1}{5}$。
2. $23+(-17)+16+(-22)=(23+16)+[(-17)+(-22)]=39-39=0$。
3. $6-(-6)=12$；$0-9=-9$；$(-\frac{1}{2})-(-\frac{1}{4})=-\frac{1}{4}$。
4. $-4+23=19$，室内温度是 $19^\circ C$。
5. 设江面上方为正，桥面为 $48$，江底为 $-10$，距离是 $48-(-10)=58$ m。
