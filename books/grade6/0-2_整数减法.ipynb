# 0.2 整数减法

减法可以直接算，也可以转成加法来理解。代码里最稳定的做法，是先弄清楚“差”和“相反数”的关系。

## 思维视角

**化归。** 化归就是把不熟悉的问题变成已经会做的问题。整数减法可以化成加法：`a - b = a + (-b)`。

```python
examples = [(8, 3), (8, -3), (-8, 3), (-8, -3)]

for a, b in examples:
    print(f"{a} - ({b}) = {a} + ({-b}) = {a + (-b)}")
```

## 在数轴上看减法

`a - b` 可以理解成：从 `a` 出发，朝着 `b` 的相反方向走。减去正数向左走，减去负数向右走。

```python
def subtract(a, b):
    return a + (-b)

print(subtract(10, 4))
print(subtract(10, -4))
print(subtract(-6, 5))
```

## 小结

整数减法的核心不是背很多规则，而是记住一条转换：减去一个数，等于加上它的相反数。
