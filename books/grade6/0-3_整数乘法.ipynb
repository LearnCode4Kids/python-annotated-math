# 0.3 整数乘法

这一节把整数乘法看成“重复加法”和“按位拆分”。竖式乘法并不是新规则，而是把一个大乘法拆成几个更小的乘法，再把部分结果加起来。

## 思维视角

**拆分 + 构造。** 拆分是把多位数按个位、十位、百位拆开；构造是把每一部分的乘积重新合成最后答案。

## 从重复加法开始

`6 * 4` 可以看成 4 个 6 相加，也可以看成 6 个 4 相加。乘法是加法的压缩写法。

```python
a = 6
b = 4

total = 0
for _ in range(b):
    total = total + a

print(total)
print(a * b)
```

## 一位数乘多位数

用列表表示整数时，`[1, 2, 3]` 表示 123。计算 `123 * 4` 时，从个位开始乘，因为个位最先产生进位。

```python
number = [1, 2, 3]
one_digit = 4

result = []
carry = 0

for index in range(len(number) - 1, -1, -1):
    digit_product = number[index] * one_digit + carry
    result.insert(0, digit_product % 10)
    carry = digit_product // 10

while carry > 0:
    result.insert(0, carry % 10)
    carry = carry // 10

print(result)
```

## 多位数乘多位数

`123 * 45` 可以拆成 `123 * 5 + 123 * 40`。竖式里“往左一位，后面补一个 0”，就是在表示乘以 10。

```python
partial_ones = 123 * 5
partial_tens = 123 * 4 * 10

print(partial_ones)
print(partial_tens)
print(partial_ones + partial_tens)
```
