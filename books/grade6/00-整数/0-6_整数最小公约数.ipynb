# 0.6 整数最大公约数

文件名沿用“整数最小公约数”的说法，但本节内容对应 `gcd`，也就是**最大公约数**。两个正整数的最大公约数，是能同时整除它们的最大整数。

## 思维视角

**枚举验证 + 记录最优。** 从 1 开始试每个可能的因数。只要它能同时整除两个数，就把它记录下来。最后留下来的就是最大的公约数。

```python
a = 12
b = 18
common_factors = []

for factor in range(1, min(a, b) + 1):
    if a % factor == 0 and b % factor == 0:
        common_factors.append(factor)

print(common_factors)
print(max(common_factors))
```

## 写成函数

函数只需要两个输入：`a` 和 `b`。循环范围到较小的那个数就够了，因为公约数不可能比两个数都大。

```python
def gcd(a, b):
    answer = 1
    smaller = min(a, b)

    for factor in range(1, smaller + 1):
        if a % factor == 0 and b % factor == 0:
            answer = factor

    return answer

print(gcd(12, 18))
print(gcd(8, 20))
print(gcd(7, 13))
```

## 第二种方法：辗转相除法

枚举法容易理解，但如果数字很大，会试很多次。辗转相除法用的是一个重要观察：`a` 和 `b` 的最大公约数，也会整除 `a` 除以 `b` 后的余数。于是问题可以变小：先看 `gcd(a, b)`，再看 `gcd(b, r)`。当余数变成 0 时，最后留下的除数就是最大公约数。

```python
a = 18
b = 12

while b != 0:
    r = a % b  # r 是 remainder 的缩写，表示余数。
    print(a, "除以", b, "余", r)
    a = b
    b = r

print("最大公约数是", a)
```

把这个过程写成函数，就不需要保存所有公约数。每次只保留当前的两个数和余数。

```python
def gcd_fast(a, b):
    while b != 0:
        r = a % b  # r 是 remainder 的缩写，表示余数。
        a = b
        b = r

    return a

print(gcd_fast(12, 18))
print(gcd_fast(8, 20))
print(gcd_fast(7, 13))
```
