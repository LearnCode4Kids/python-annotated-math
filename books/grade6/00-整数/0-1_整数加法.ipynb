# 0.1 整数加法

这一节把小学竖式加法改写成 Python 代码。一个多位数可以看成一串数字，从个位开始处理进位。

## 思维视角

**拆分 + 逆向观察。** 竖式加法从右往左算，因为个位最先产生进位。代码里也从列表的最后一位开始循环。

## 用列表表示整数

例如 `356` 可以写成 `[3, 5, 6]`。列表里的每个数字都在 0 到 9 之间。

```python
number = [3, 5, 6]

value = 0
for digit in number:
    value = value * 10 + digit

print(value)
```

## 逐位相加

先看长度相同的情况。关键变量是 `carry`，它表示从右边一位传来的进位。

```python
left = [3, 5, 6]
right = [4, 7, 8]

result = []
carry = 0

for index in range(len(left) - 1, -1, -1):
    digit_sum = left[index] + right[index] + carry
    result.insert(0, digit_sum % 10)
    carry = digit_sum // 10

if carry > 0:
    result.insert(0, carry)

print(result)
```

## 长度不同怎么办

`56 + 478` 可以先看成 `056 + 478`。短的列表前面补 0，数值不变，但位置对齐了。

```python
left = [5, 6]
right = [4, 7, 8]

max_len = max(len(left), len(right))
left = [0] * (max_len - len(left)) + left
right = [0] * (max_len - len(right)) + right

print(left)
print(right)
```
