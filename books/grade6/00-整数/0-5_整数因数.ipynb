# 0.5 整数因数

这一节从乘法开始看因数。能整除一个数的整数，就是它的因数。质数和质因数分解，都是在研究“一个数由哪些乘法零件组成”。

## 思维视角

**枚举验证 + 分解。** 枚举就是一个一个试；分解就是把一个大数拆成更小、更基础的数。

## 乘法里的因数

如果 `a * b = c`，那么 `a` 和 `b` 都是 `c` 的因数。

```python
number = 24
factors = []

for candidate in range(1, number + 1):
    if number % candidate == 0:
        factors.append(candidate)

print(factors)
```

## 质数

质数只有两个正因数：1 和它自己。判断质数时，可以从 2 开始试除。只要找到一个能整除的数，它就不是质数。

```python
def is_prime(x):
    if x < 2:
        return False

    for factor in range(2, x):
        if x % factor == 0:
            return False

    return True

for x in range(1, 11):
    print(x, is_prime(x))
```

## 质因数分解

把一个数不断除以能整除它的最小质数，就能得到从小到大的质因数列表。

```python
def num_to_factors(x):
    factors = []
    divisor = 2

    while x > 1:
        if x % divisor == 0:
            factors.append(divisor)
            x = x // divisor
        else:
            divisor += 1

    return factors

print(num_to_factors(12))
print(num_to_factors(18))
```
