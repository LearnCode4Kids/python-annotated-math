# 0.6 整数最大公约数

文件名沿用“整数最小公约数”的说法，但本节内容对应 `gcd`，也就是**最大公约数**。两个正整数的最大公约数，是能同时整除它们的最大整数。

## 思维视角

**枚举验证 + 记录最优。** 从 1 开始试每个可能的因数。只要它能同时整除两个数，就把它记录下来。最后留下来的就是最大的公约数。

```python
a = 12
b = 18
common_factors = []

for candidate in range(1, min(a, b) + 1):
    if a % candidate == 0 and b % candidate == 0:
        common_factors.append(candidate)

print(common_factors)
print(max(common_factors))
```

## 写成函数

函数只需要两个输入：`a` 和 `b`。循环范围到较小的那个数就够了，因为公约数不可能比两个数都大。

```python
def gcd(a, b):
    answer = 1

    for candidate in range(1, min(a, b) + 1):
        if a % candidate == 0 and b % candidate == 0:
            answer = candidate

    return answer

print(gcd(12, 18))
print(gcd(8, 20))
print(gcd(7, 13))
```
