# fractions 库：用 Fraction 准确表示分数

本小节只介绍 1.1 有理数的引入中用到的 `Fraction`。它来自 Python 标准库 `fractions`，不用额外安装。

## 为什么需要它

数学课里经常写 $\frac{1}{2}$、$-\frac{7}{4}$。如果直接写成小数，学生容易看不出它原来的分数结构。

`Fraction` 可以让 Python 用分数的形式保存这些数。

```python
from fractions import Fraction

a = Fraction(7, 4)
b = Fraction(-1, 2)

print(a)
print(b)
print(a + b)
```

## 本课只用到的写法

- `Fraction(7, 4)` 表示 $\frac{7}{4}$。
- `Fraction(-1, 2)` 表示 $-\frac{1}{2}$。
- `Fraction(3, 1)` 表示整数 $3$，也就是 $\frac{3}{1}$。

注意：第二个数是分母，不能为 0。

```python
numbers = [Fraction(3, 1), Fraction(1, 2), Fraction(-7, 4), Fraction(0, 1)]

for x in numbers:
    print(f"{x} = {x.numerator}/{x.denominator}")
```

## 小结

在本课程中，`Fraction` 的作用很简单：**让分数保持分数样子**。这样画数轴、判断正负、观察相反数时，表达会更接近教材。
