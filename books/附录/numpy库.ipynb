# NumPy 库：生成数轴刻度

本小节只介绍 1.1 有理数的引入中用到的 `np.arange(...)`。

## 为什么需要它

画数轴时，我们需要一串等间隔的刻度，比如：

$$-4, -3, -2, -1, 0, 1, 2, 3, 4$$

`numpy` 可以帮我们快速生成这样的数字序列。

```python
import numpy as np

ticks = np.arange(-4, 5)
print(ticks)
```

## 本课只用到的写法

`np.arange(start, stop)` 会生成从 `start` 开始，到 `stop` 之前结束的一串数。

所以 `np.arange(-4, 5)` 会得到 -4 到 4，而不会包含 5。

```python
xmin = -3.2
xmax = 4.7

ticks = np.arange(np.ceil(xmin), np.floor(xmax) + 1)
print(ticks)
```
