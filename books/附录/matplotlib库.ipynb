# Matplotlib 库：画出数轴

本小节只介绍 1.1 有理数的引入中用到的 `matplotlib.pyplot`。它在代码里通常简写成 `plt`。

## 为什么需要它

有理数、数轴、相反数都很适合画出来看。`matplotlib` 的作用就是帮我们把“数”和“点的位置”变成图形。

```python
import os
os.environ.setdefault("MPLCONFIGDIR", "/tmp/matplotlib")
os.environ.setdefault("XDG_CACHE_HOME", "/tmp")

import matplotlib.pyplot as plt

plt.rcParams["axes.unicode_minus"] = False
```

## 本课只用到的几个动作

- `plt.subplots(...)`：准备一张图和一块画布。
- `ax.plot(...)`：画线段或刻度。
- `ax.scatter(...)`：画点。
- `ax.text(...)`：写文字。
- `plt.show()`：显示图像。

```python
fig, ax = plt.subplots(figsize=(7, 1.6))

axis_right = 4.35
ax.annotate("", xy=(axis_right, 0), xytext=(-4, 0), arrowprops=dict(arrowstyle="->", color="black"))

for t in range(-4, 5):
    ax.plot([t, t], [-0.06, 0.06], color="black")
    ax.text(t, -0.18, str(t), ha="center", va="top")

ax.scatter([-3, 3], [0, 0], color="#d62728", s=70)
ax.text(-3, 0.22, "-3", ha="center", color="#d62728")
ax.text(3, 0.22, "3", ha="center", color="#d62728")

ax.set_xlim(-4.4, axis_right + 0.05)
ax.set_ylim(-0.5, 0.6)
ax.set_title("Opposite numbers on a number line")
ax.axis("off")
plt.show()
```
